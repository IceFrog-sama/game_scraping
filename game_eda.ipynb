```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

score_data_valve = valve_df[['Name', 'Metascore', 'User Score']]
score_data_blizzard = blizzard_df[['Name', 'Metascore', 'User Score']]

score_data_valve['Game Company'] = 'Valve'
score_data_blizzard['Game Company'] = 'Blizzard'
# create boolean missing in case we need it for color coding later
#score_data['Missing Metascore'] = [np.isnan(x) for x in score_data['Metascore']]
score_data = pd.concat([score_data_valve, score_data_blizzard], ignore_index=True)
score_data
```

```python
# drop if there's more than 1 NaN in the subset (i.e. both metascore and userscore are NaN)
score_data = score_data.dropna(subset=['Metascore', 'User Score'], thresh=1)
score_data = score_data.reset_index(drop=True)
score_data
```

```python
score_data['Metascore filled'] = [uscore * 10 if np.isnan(score_data['Metascore'][i]) 
                    else score_data['Metascore'][i] for i, uscore in enumerate(score_data['User Score'])]
#score_data
```

```python
data_for_cluster = score_data[['Metascore filled', 'User Score']]

kmeans = KMeans(n_clusters=4, random_state=0).fit(data_for_cluster)
clusters = kmeans.predict(data_for_cluster)
score_data['Cluster'] = clusters
score_data
```

```python
valve_games = score_data[score_data['Game Company'] == 'Valve']
blizzard_games = score_data[score_data['Game Company'] == 'Blizzard']
#valve_scores = valve_games[['Metascore filled', 'User Score']]
#blizzard_scores = blizzard_games[['Metascore filled', 'User Score']]
valve_games.head()
```

```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import requests

# get mask
url='https://www.pcgamesn.com/wp-content/uploads/2018/09/steam-logo.jpg'
# alt: https://www.geeky-gadgets.com/wp-content/uploads/2012/08/Valve-Games.jpg
im = requests.get(url, stream=True).raw
steam_mask = np.array(Image.open(im))

# generate word cloud
wc = WordCloud(background_color="white", max_words=100, mask=steam_mask,
               max_font_size=90, random_state=42)
wc.generate(' '.join(valve_games['Name']))

# create coloring from image
image_colors = ImageColorGenerator(steam_mask)
plt.figure(figsize=[15,15])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```

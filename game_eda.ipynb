### Which company has better ratings?

#### Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("colorblind")
import numpy as np

score_data_valve = valve_df[['Name', 'Metascore', 'User Score', 'Release']]
score_data_blizzard = blizzard_df[['Name', 'Metascore', 'User Score', 'Release']]

score_data_valve['Game Company'] = 'Valve'
score_data_blizzard['Game Company'] = 'Blizzard'
score_data = pd.concat([score_data_valve, score_data_blizzard], ignore_index=True)
score_data.tail()
```

```python
# drop if there's more than 1 NaN in the subset (i.e. both metascore and userscore are NaN)
score_data = score_data.dropna(subset=['Metascore', 'User Score'], thresh=1)
score_data = score_data.reset_index(drop=True)
score_data.tail()
```

```python
# Impute missing metascores with user scores
score_data['Metascore filled'] = [uscore * 10 if np.isnan(score_data['Metascore'][i]) 
                    else score_data['Metascore'][i] for i, uscore in enumerate(score_data['User Score'])]
```

```python
data_for_cluster = score_data[['Metascore filled', 'User Score']]

# create clusters based on metascore and user score
kmeans = KMeans(n_clusters=4, random_state=0).fit(data_for_cluster)
clusters = kmeans.predict(data_for_cluster)
score_data['Cluster'] = clusters
score_data
```

```python
valve_games = score_data[score_data['Game Company'] == 'Valve']
blizzard_games = score_data[score_data['Game Company'] == 'Blizzard']
valve_games.head()
```

```python
# Clustered scores (not separated by game company)

sns.scatterplot(
    x=score_data['Metascore filled'],
    y=score_data['User Score'],
    hue=score_data['Cluster'].astype("category") # Set color according to clusters, categorically
)
plt.xlabel('Metascore')
plt.ylabel('Userscore')
```

```python
# Clustered scores (marker separated by game company)

sns.scatterplot(
    x=score_data['Metascore filled'],
    y=score_data['User Score'],
    hue=score_data['Cluster'].astype("category"),
    style=score_data['Game Company'] # Set shape according to company
)

plt.xlabel('Metascore')
plt.ylabel('User score')
```

#### Comparison by Mean/Median Scores

```python
valve_games['Metascore filled'].mean(), blizzard_games['Metascore filled'].mean()
```

```python
valve_games['User Score'].mean(), blizzard_games['User Score'].mean()
```

Insights:
- Based on Clustering:
    - Valve game ratings in all clusters from low metascore, user score to high metascore, user score
    - Blizzard games exist in cluster 2 and 3 where games have consistently high metascore with middle to high user score
- Based on Mean/Median:
    - Valve games have a higher average user score than Blizzard
    - Blizzard games have a higher metascore than Valve
- Reasoning:
    - Metascores are only influenced by a select group of renowned critics and usually given to a game upon release
    - User scores can be from anyone on the internet
- Conclusion:
    - Blizzard games may be better on release but poorly maintained: causing user scores to go down over time
    - Blizzard games may have better connections with critics to drive metascores up (controversial)
    - Valve games may be more consistent in maintaining their games to keep user scores high

### Do ratings improve over time?

```python
# Interesting - 4 games were released in 10/10/2007
valve_games.sort_values(by='Release',inplace=True)
valve_games
```

```python
blizzard_games.sort_values(by='Release', inplace=True)
blizzard_games
```

```python
# Pivot data to long form 

valve_long = valve_games[["User Score", "Release", "Metascore filled"]].melt(
    id_vars=("Release"), value_name="Score", var_name="Score Type"
)

# Multiply Use scores by 10 to have a common scale
valve_long.loc[valve_long["Score Type"] == "User Score", "Score"] *= 10
```

```python
sns.lineplot(
    x=valve_long["Release"],
    y=valve_long["Score"],
    hue=valve_long["Score Type"],
    style=valve_long["Score Type"]
)

plt.title('Valve Game Ratings Over Time', fontsize=15)
plt.xlabel('Release Date')
plt.ylabel('Scores')
plt.xticks(rotation=30, ha='right')
```

```python
blizzard_long = blizzard_games[["User Score", "Release", "Metascore filled"]].melt(
    id_vars=("Release"), value_name="Score", var_name="Score Type"
)
blizzard_long.loc[blizzard_long["Score Type"] == "User Score", "Score"] *= 10
```

```python
sns.lineplot(
    x=blizzard_long["Release"],
    y=blizzard_long["Score"],
    hue=blizzard_long["Score Type"],
    style=blizzard_long["Score Type"]
)

plt.title('Blizzard Game Ratings Over Time', fontsize=15)
plt.xlabel('Release Date')
plt.ylabel('Scores')
plt.xticks(rotation=30, ha='right')
```

Insights:
- Valve ratings: 
    - User score and metascore have been fairly consistent
    - Except for 2 failed games around 2014 and 2019 where user score plummeted
- Blizzard ratings:
    - User score and metascore were consistent up until 2004
    - Then user score started being consistently lower than metascore (with increasing gaps)
